```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
textimage1 = cv2.imread('TextImage.jpg', 0)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(textimage1, cmap = 'gray')
plt.show()
```

# Dilation

```python
structural_element = np.ones((5,5),np.uint8)
dilated = cv2.dilate(textimage1, structural_element, iterations = 1) ## iteration = 1 means one pass of kernel and we can see the connection
                                                     # boundary between A and B has eroding away. More effect is observed with 
                                                     # highest value iterations
```

```python
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.imshow(textimage1, cmap = 'gray')
plt.title('Input Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(dilated, cmap = 'gray')
plt.title('Image After Dilation', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
textimage2 = cv2.imread('text.jpg', 0)
```

```python
structural_element = np.ones((9,9),np.uint8)
dilated2 = cv2.dilate(textimage2, structural_element, iterations = 1)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(textimage2, cmap = 'gray')
plt.show()
```

```python
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.imshow(textimage2, cmap = 'gray')
plt.title('Input Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(dilated2, cmap = 'gray')
plt.title('Image After Dilation', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.show()
```

# Dilation With Rectangular Structural Element

# Erosion

## Erodes away boundaries of foreground objects. Works best when foreground is light color (preferrably white) and background is dark.

```python
plt.figure(figsize=(10,6))
plt.imshow(textimage1, cmap = 'gray')
plt.show()
```

```python
structural_element = np.ones((5,5),np.uint8)
eroded = cv2.erode(textimage1,structural_element,iterations = 1) 
```

# Morphological Filtering

 # Opening


```python
plt.figure(figsize=(10,6))
plt.imshow(textimage1, cmap = 'gray')
plt.show()
```

```python
Noise = np.random.randint(low = 0, high = 2 ,size = textimage1.shape, dtype = 'uint8')
```

```python
Noise
```

```python
noisyImage = textimage1 + whiteNoise
```

```python
plt.figure(figsize=(10,6))
plt.imshow(whiteNoise, cmap = 'gray')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.imshow(noisyImage, cmap = 'gray')
plt.show()
```

```python
textimage1.dtype
```

```python
noisyImage.dtype
```

```python
opening = cv2.morphologyEx(noisyImage, cv2.MORPH_OPEN, structural_element)
```

```python
plt.figure(figsize = (16,8))
plt.subplot(131)
plt.imshow(textimage1, cmap = 'gray')
plt.title('Original Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(132)
plt.imshow(noisyImage, cmap = 'gray')
plt.title('Noisy Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(133)
plt.imshow(opening, cmap = 'gray')
plt.title('Image After Opening', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.show()
```

# Closing


```python
plt.figure(figsize=(10,6))
plt.imshow(textimage1, cmap = 'gray')
plt.show()
```

```python
blackNoise = Noise * (-255) # multiply by -255 to create black noise
```

```python
plt.figure(figsize=(10,6))
plt.imshow(blackNoise, cmap = 'gray')
plt.show()
```

```python
noisyImage_black = textimage1 + blackNoise
```

```python
plt.figure(figsize=(10,6))
plt.imshow(noisyImage_black, cmap = 'gray')
plt.show()
```

```python
closing = cv2.morphologyEx(noisyImage_black, cv2.MORPH_CLOSE, structural_element)
```

```python
plt.figure(figsize = (16,8))
plt.subplot(131)
plt.imshow(textimage1, cmap = 'gray')
plt.title('Original Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(132)
plt.imshow(noisyImage_black, cmap = 'gray')
plt.title('Noisy Image', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.subplot(133)
plt.imshow(closing, cmap = 'gray')
plt.title('Image After closing', fontsize = 15)
plt.xticks([])
plt.yticks([])
plt.show()
```
