```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Function to display images
def show_image(image, title):
    plt.figure(figsize=(4, 4))
    plt.imshow(image, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()

# 1. Create black images (grayscale and color)
gray_image = np.zeros((224, 224), dtype=np.uint8)
color_image = np.zeros((224, 224, 3), dtype=np.uint8)
show_image(gray_image, "Grayscale Image")
show_image(color_image, "Color Image")
```

```python

# 2. Draw lines
cv2.line(gray_image, (10, 10), (200, 10), (255), 2)
cv2.line(color_image, (10, 30), (200, 30), (0, 255, 0), 2)
show_image(gray_image, "Grayscale Image with Line")
show_image(color_image, "Color Image with Line")
```

```python

# 3. Draw rectangles
cv2.rectangle(gray_image, (50, 50), (150, 100), (200), 2)
cv2.rectangle(color_image, (50, 120), (150, 170), (255, 0, 0), 2)
show_image(gray_image, "Grayscale Image with Rectangle")
show_image(color_image, "Color Image with Rectangle")
```

```python

# 4. Draw circles
cv2.circle(gray_image, (112, 150), 30, (180), 2)
cv2.circle(color_image, (112, 200), 30, (0, 0, 255), 2)
show_image(gray_image, "Grayscale Image with Circle")
show_image(color_image, "Color Image with Circle")
```

```python

# 5. Draw polygons
pts = np.array([[50, 50], [170, 50], [190, 120], [30, 150]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(gray_image, [pts], isClosed=True, color=(150), thickness=2)
cv2.polylines(color_image, [pts], isClosed=True, color=(255, 255, 0), thickness=2)
show_image(gray_image, "Grayscale Image with Polygon")
show_image(color_image, "Color Image with Polygon")
```

```python

# 6. Write text
cv2.putText(gray_image, "Image Processing", (10, 200), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255), 1)
cv2.putText(color_image, "Computer Vision", (10, 220), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)
show_image(gray_image, "Grayscale Image with Text")
show_image(color_image, "Color Image with Text")
```
